```python
# --- Disease Spread Simulation (SIR Model) ---

import numpy as np
import matplotlib.pyplot as plt

# Total population
N = 10000

# Initial values
I0 = 10      # initially infected
R0 = 0       # initially recovered
S0 = N - I0 - R0  # initially susceptible

# Contact rate (beta) and recovery rate (gamma)
beta = 0.3      # infection rate
gamma = 0.1     # recovery rate

# Time steps (days)
days = 160
t = np.linspace(0, days, days)

# Arrays to store S, I, R values
S = np.zeros(days)
I = np.zeros(days)
R = np.zeros(days)

# Initial data
S[0] = S0
I[0] = I0
R[0] = R0

# SIR equations simulation
for day in range(1, days):
    dS = -beta * S[day-1] * I[day-1] / N
    dI = beta * S[day-1] * I[day-1] / N - gamma * I[day-1]
    dR = gamma * I[day-1]

    S[day] = S[day-1] + dS
    I[day] = I[day-1] + dI
    R[day] = R[day-1] + dR

# Plotting
plt.figure(figsize=(10,6))
plt.plot(t, S, label='Susceptible')
plt.plot(t, I, label='Infected')
plt.plot(t, R, label='Recovered')

plt.xlabel('Days')
plt.ylabel('Number of People')
plt.title('Disease Spread Simulation (SIR Model)')
plt.grid(True)
plt.legend()
plt.show()
```
